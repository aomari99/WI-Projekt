```python
#import librarys
import pandas as pd
import numpy as np
```

```python
frame = pd.read_csv(r'weekcount.csv')
df = pd.DataFrame(frame)
```

```python
#get weather data
wframe = pd.read_csv(r'weather.csv')
weather = pd.DataFrame(wframe)
weather.head()
```

```python
df = pd.merge(df,weather,on=['year','week'],how='outer').dropna()
#df = df.reset_index()
print(df[df.index.duplicated()])

df
```

```python
df.plot.scatter(x='anzahl',
                      y='precipMM'
                      
                      )
```

```python
p = df.plot.scatter(x='anzahl',
                      y='sunHour'
                      
                      )
p.get_figure().savefig("anzal_sunhour.jpg",dpi=300)
```
