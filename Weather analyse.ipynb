```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#get weather data
wframe = pd.read_csv(r'weather.csv')
weather = pd.DataFrame(wframe)
weather.head()
```

```python
import seaborn as sns
sns_plot  = sns.pairplot(weather, diag_kind='kde')
sns_plot.savefig("weatherpair.jpg",dpi=300)
```
