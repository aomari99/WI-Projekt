```python
import folium
from folium.features import DivIcon
import pandas as pd 
```

```python
df = pd.read_csv(r'D:/Projects/CSVDataming/geoyear.csv')
```

```python
weight = df['2019']
lats = df['lat1']
longs = df['long1']
names = df['station1']
lat = df['lat1'].mean()
long = df['long1'].mean()
print(lat)
print(long)
```

```python
m = folium.Map(location=[lat, long],zoom_start=14)
```

```python
for i in range (len(lats)):
    folium.CircleMarker(
    location=[lats[i], longs[i]],
    radius=weight[i]/1000,
    popup=names[i],
    color='#ff0000',
    fill=True,
    fill_color='#ff0000'
).add_to(m)
    #folium.map.Marker(  [lats[i]   , longs[i]   ], icon=DivIcon(  icon_size=(120,200),      icon_anchor=(60,0),       html='<div style="font-size: 12pt;color:blue">%s</div>' % names[i],       )   ).add_to(m)
```
